```python
import numpy as np
from  keras.datasets import imdb #movie review datasets 
```

```python
top_words = 5000  #(top 5000 review remain blank)
(x_train,y_train),(x_test, y_test) = imdb.load_data(num_words=top_words)  #unsing ferquently words for remove the misspelled word
```

```python
# preprocessing 
# going to fix the lenght of eahch review
from keras.preprocessing import sequence
# total no of reviews = 25k
max_review_length = 500
X_train = sequence.pad_sequences(x_train, maxlen=max_review_length)
X_test = sequence.pad_sequences(x_test, maxlen=max_review_length)

print(X_train.shape)
print(X_test.shape)
```

```python
print(X_train[0])   #its add pdding default whay -- > better result , 
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.layers.embeddings import Embedding  #first layesr the convert the layer into vactors

```

```python
embedding_vectore_length = 32

model = Sequential()

model.add(Embedding(top_words,  #5000
                   embedding_vectore_length,  #32
                   input_length = max_review_length))  #500-->input shape 
model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2)) #output length of vectore 
model.add(Dense(1,activation='sigmoid'))


model.compile(loss="binary_crossentropy",
             optimizer = "adam",
             metrics = ["accuracy"])
model.summary()
```

```python
model.fit(X_train , y_train,
         epochs=3,
         batch_size=64)  #aftre 64 review in time will one time weight update 
```

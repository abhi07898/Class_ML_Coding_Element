```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline

import cv2
```

```python
im = cv2.imread('./img/tree.jpg')
im.shape  #height rwo, width, 3 color image 
```

```python
img  = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
plt.show()   #stride =1 it may more than 1 also 
```

```python
h, w= img.shape  #there is no c bcs we convert it into gray
print(h,w)
```

```python
# K = np.array([
#     [0,-1,0],
#     [-1,4,-1],
#     [0,-1,0]
# ])    #craete filter by hardcode filter also known as kernal 

#Sobel Filter: Horizantal edges detect
# K = np.array([
#     [-1,-2,-1],
#     [0,0,0],
#     [1,2,1]
# ])  

#sabel filter : verticalyy
# K = np.array([
#     [-1,0,1],
#     [-2,0,2],
#     [-1,0,1]
# ])  


#blur
K = np.array([
    [.1,.1,.1],
    [.1,.1,.1],
    [.1,.1,.1]
]) 
```

```python
kh, kw = 3,3   #kernal width 
```

```python
oh , ow  = h-kh+1, w-kw+1   #out heght out width
```

```python
# empty image creation 
outImg = np.zeros((oh, ow))
```

```python
for hi in range(oh):
    for wi in range(ow):
        patch =img[hi:hi+kh,wi:wi+kw ] # decide pathch means 0 to 2,1to 3, 2 to 4 
#         applying filter 
        prod = patch* K 
        prod = prod.sum()
        outImg[hi , wi] = prod if prod > 0 else 0
```

```python
plt.imshow(outImg, cmap= 'gray')
plt.show()
```

# REstructing Data

```python
weather = pd.read_csv('./DataSet/weather/weather.csv')
weather.head()
```

# melt 

```python
weather.shape
```

```python
weather_melt = pd.melt(weather, 
        id_vars=['year', 'month', 'element'],
       var_name = 'day',
       value_name='temp')
```

```python
weather_melt
```

```python
weather_melt.shape
```

Restruct Pivot 

```python
weather_melt.pivot_table(index=['year','month','day'],
                        columns = 'element',
                        values='temp')
```

```python
weather_melt.pivot_table(index=['year','month','day'],
                        columns = 'element',
                        values='temp').reset_index()
```

```python
weather_tidy = weather_melt.pivot_table(index=['year','month','day'],
                        columns = 'element',
                        values='temp',
                        dropna = False).reset_index()
```

```python
weather_tidy.head()
```

# saving the data to a file 

```python
weather_tidy.to_csv('output/weather_tidy.csv')
```

```python
weather_tidy.to_csv('output/weather_tidy.csv')
```

```python
x = np.arange(1,10)  #for black = k
y = x**2

plt.title("Quadratics")
plt.xlabel("X")
plt.ylabel("X^2")
plt.grid()  # for grid view

plt.text(8,60,"hello")


plt.xlim(-2,12)
plt.ylim(-10,100)

plt.plot(x,y,'g--', label = 'Squares', linewidth = 4) #for combining color and shapes (ko , rv, b*)
plt.legend() #label does not work until legend is not define
plt.show()
```

```python
import math
X = np.arange(1,101)
Y =  [math.log(v) for v in X]
plt.plot(X,Y)
plt.show()
```

```python
X = np.arange(1,101)
Y =  np.log(X)
plt.plot(X,Y)
plt.show()
```

```python
X = np.arange(1,101)
Y =  list(map(math.log,X))
plt.plot(X,Y)
plt.show()
```

```python
mu, sigma = 100, 15
S = np.random.normal(mu,sigma,1000)

print(np.mean(S))
print(np.std(S))
```

```python
plt.hist(S, 20)  #hist(S, bucket) this bucket by default is divided by 10 
plt.show()
```

```python
def sigmoid(z):
    return 1/(1+math.exp(-z))
```

```python
X = np.arange(-10,11)
Y = [sigmoid(v) for v in X]

plt.plot(X,Y)
plt.show()
```

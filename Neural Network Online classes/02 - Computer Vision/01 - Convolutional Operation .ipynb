```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2
```

```python
im = cv2.imread("./img/tree.jpg")   #image reading 
img = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)  # converting the im into BGR2GRAy
plt.imshow(img, cmap='gray')   #plotting tghe image..  
plt.show()      #show the image 
```

```python
print(im.shape)  #color image where dimension = 3
print(img.shape)  # gray image whree dim = 1
```

# FILTERS hardcoded 

```python
#sharpen 
K = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0 , -1, 0]
])

print(K.shape)  #  3x3
outR = img.shape[0] - K.shape[0] + 1 # i-j+1
outC = img.shape[1] - K.shape[1] + 1  # smae here 

outImg = np.zeros((outR, outC))  # starting all values should  0 bcs 0  mattrix
print(outImg.shape)




for r in range(outR):
    for c in range(outC):
        imgPatch = img[r:r+3,c:c+3]
        out = imgPatch * K
        outImg[r,c] = out.sum() if out.sum() >= 0 else 0
plt.imshow(outImg, cmap='gray')

    
```

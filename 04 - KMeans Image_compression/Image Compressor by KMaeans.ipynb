```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2

plt.rcParams['figure.figsize']=[15,10]
```

```python
im = cv2.imread('img/nature.jpg')
im.shape
```

```python
(h,w,c) = im.shape
(h,w,c)
```

```python
image=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
plt.imshow(np.hstack([im,image]))
plt.show()
```

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
flatImg = image.reshape(h*w,3)
flatImg.shape
```

```python
clt = MiniBatchKMeans(n_clusters=16)
labels = clt.fit_predict(flatImg)
labels.shape
```

```python
clt.cluster_centers_.astype('uint8')
```

```python
quants = clt.cluster_centers_.astype('uint8')[labels]
```

```python

quants.shape
```

```python
resultImg = quants.reshape(h,w,3)
plt.imshow(resultImg)
plt.show
```

```python
#back and save to thios iomage into a file
```

```python
result= cv2.cvtColor(resultImg, cv2.COLOR_RGB2BGR)
#plt.imshow(result)
cv2.imwrite(r'output/compressed.jpg',result)
```
